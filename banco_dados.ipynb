```python
#Conexao com bd
import psycopg2

def conecta_db():
    con = psycopg2.connect(host='localhost', 
                         database='companies',
                         user='postgres', 
                         password='*')
    return con
```

```python
#insert
def inserir_db(sql):
    con = conecta_db()
    cur = con.cursor()
    try:
        cur.execute(sql)
        con.commit()
    except (Exception, psycopg2.DatabaseError) as error:
        print("Error: %s" % error)
        con.rollback()
        cur.close()
        return 1
    cur.close()
```

```python
#read
def read_db(sql):
    con = conecta_db()
    cur = con.cursor()
    cur.execute(sql)
    res_values = cur.fetchall() # retorna os valores
    res_columns = cur.description #retorna as colunas
    cur.close()
    return res_values, res_columns
```

```python
sql = ('select * from companies')
```

```python
df_values, df_columns = read_db(sql)
```
